```python
#import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To view the matplotlib in notebook
%matplotlib inline
```

```python
#import the auto dataset
Auto_DF = pd.read_csv('auto_data.csv')
```

```python
#view first 5 records
Auto_DF.head()
```

```python
#write a user defined function for origin
def origin(num):
  if num ==1:
    return 'USA'
  elif num==2:
    return 'Europe'
  else:
    return 'Asia'
```

```python
#Use apply function on data set
Auto_DF['origin']=Auto_DF['origin'].apply(origin)
```

```python
#view first 30 records
Auto_DF.head(30)
```

```python
#Draw the pairplot using seaborn for mpg, weight and origin
sns.pairplot(Auto_DF[['mpg','weight','origin']],hue='origin',size=8)
```
